1) How many customers have placed single orders and multiple orders.

```python
#1)
customer_order_counts = df['Store Order Id'].value_counts()
multiple_orders_customers = customer_order_counts[customer_order_counts > 1].count()

total_orders = len(df)
single_order_customers = len(customer_order_counts) - multiple_orders_customers
multiple_order_customers = multiple_orders_customers

print("Total Orders:", total_orders)
print("Single Order Customers:", single_order_customers)
print("Multiple Order Customers:", multiple_order_customers)
```

1) top-selling products by quantity ordered described with bar chart.

```python
#1)
top_selling_products = df.groupby('Style ID')['Quantity'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 8))
top_selling_products.plot(kind='bar')
plt.title('Top 10 Selling Products by Quantity Ordered')
plt.xlabel('Style ID')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
